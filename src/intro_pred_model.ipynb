# Intro to using Prediction Model Class
Prediction model is a class used to train an RNN for stock prediction. Thus far we have 500 stocks which are used to train the model. This class also can be used to easily test new data.

```python
import quandl
import matplotlib.pyplot as plt
import prediction_model
```

Creating a prediction model object which will create a financial data object with the training and testing data

```python
pm=prediction_model.prediction_model()
```

To train the model one can use the train_data method, which initializes the training sequence. Various parameters can be tuned in the constructor level of the class 

```python
pm.train_data()
```

Data can easily be tested then by grabbing stocks from the wiki data. Then the build in method test stock can be used to get the resulting values y_res and actual values y_test. Additionally accuracy is returned for [.01, .1, 1, 3, 5, 7, 10] % error percentiles

```python
data=pm.data.get_data('Wiki', {'Stock':'WSO'})
acc, y_res, y_test=pm.test_stock(data)
```

Then you can easily plot the data to see the results using matplotlib. Note the results are normalized for easier training of the neural network. 

```python
pred_plot,=plt.plot(y_res.data.numpy()[100:200], label='Predicted Data')
act_plot,=plt.plot(y_test[100:200], label='Actual Data')
plt.xlabel('Days')
plt.ylabel('Normalized Price')
plt.title('WSO Stock Prediction')
plt.legend([pred_plot, act_plot])
```

Inside the prediction model class there is a financial data object. Within that object one variable is data_ls which contains a pandas dataframe of one stock. To access other stocks change the indexed position.

```python
pm.data.data_ls[1]
```

Norm_data_ls is data_ls except the columns open, high, low and close have been normalized.

X test is an array of inputs for a given y test output. Each example of x test is the input size of the rnn long. Each sample is stored in one indexed position of the list x test. These would be the previous values used to predict the next value of stock price

```python
pm.data.x_test
```

Y test is the output for a given x test input. This value is the predicted of the next stock price
